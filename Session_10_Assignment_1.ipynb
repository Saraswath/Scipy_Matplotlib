## Scipy

```python
import numpy as np
t_max=np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
t_min=np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
```

```python
import matplotlib.pyplot as pl
months = np.arange(12)
pl.plot(months, t_max, 'bo')
pl.plot(months, t_min, 'go')
pl.xlabel('Month')
pl.ylabel('Min and max temperature')
```

## 1. Fitting it to the periodic function

```python
from scipy import optimize
def periodic_t(times, avg, ampl, time_offset):
    return (avg+ ampl * np.cos((times + time_offset) * 1.8 * np.pi / times.max()))

r_max, c_max = optimize.curve_fit(periodic_t, months,t_max, [40, 20, 0])
r_min, c_min = optimize.curve_fit(periodic_t, months,t_min, [-40, 20, 0])
```

```python
days = np.linspace(0, 12, num=365)

pl.figure()
pl.plot(months, t_max, 'ro')
pl.plot(days, periodic_t(days, *r_max), 'r-')
pl.plot(months, t_min, 'bo')
pl.plot(days, periodic_t(days, *r_min), 'b-')
pl.xlabel('Month')
pl.ylabel('Temperature ($^\circ$C)')

pl.show()
```

## Matplotlib

```python
import pandas as pd
```

```python
titanic = pd.read_csv("https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv")
```

## 1. Create a pie chart presenting the male/female proportion

```python
fig,ax=pl.subplots()
prop=titanic.groupby("sex").pclass.count()
explode=(0,0.1)
ax.pie(prop,explode=explode,labels=prop.index,autopct='%1.1f%%',shadow=True,startangle=90)
ax.axis('equal')
```

## 2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
pl.scatter(titanic['age'],titanic['fare'],c=titanic["sex"].map({'male':0,'female':1}),cmap="coolwarm_r")
```
